### Splitting the data and normalization

```python
from sklearn.model_selection import train_test_split
```

```python
y = temp_df["Weight"]
X = temp_df.drop("Weight", axis=1)
```

```python
### Normalization
i = 0
for key, column in X.items():
    mean = column.mean()
    std = column.std()
    X[key] = (X[key] - mean) / std
    i = i + 1

    if i > 4:
        break
```

```python
for key, column in X.items():
    print(f"{key}: Mean = {column.mean()} Standard Deviation = {column.std()}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=67)
```

```python
X_train, X_test, y_train, y_test
```

## Training

```python
from sklearn.ensemble import (
    RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor,
    ExtraTreesRegressor, VotingRegressor, StackingRegressor
)
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
# Defining Base models

models = {
    "LinearRegression": LinearRegression(),
    "Ridge": Ridge(max_iter=10000),
    "Lasso": Lasso(max_iter=10000),
    "DecisionTree": DecisionTreeRegressor(random_state=42),
    "RandomForest": RandomForestRegressor(n_estimators=200, random_state=42),
    "ExtraTrees": ExtraTreesRegressor(n_estimators=200, random_state=42),
    "AdaBoost": AdaBoostRegressor(n_estimators=200, random_state=42),
    "GradientBoosting": GradientBoostingRegressor(n_estimators=200, random_state=42),
    "SVR": make_pipeline(StandardScaler(), SVR(kernel="rbf", C=100)),
    "KNN": make_pipeline(StandardScaler(), KNeighborsRegressor(n_neighbors=5))
}
```

```python
# Individual results

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    results[name] = {
        "MAE": mean_absolute_error(y_test, preds),
        "RMSE": np.sqrt(mean_squared_error(y_test, preds)),
        "R2": r2_score(y_test, preds)
    }
```

```python
# Voting method (simple esembling)

voting_reg = VotingRegressor([
    ('rf', models["RandomForest"]),
    ('gb', models["GradientBoosting"]),
    ('et', models["ExtraTrees"]),
    ('ridge', models["Ridge"]),
    ('svr', models["SVR"])
])

voting_reg.fit(X_train, y_train)
voting_preds = voting_reg.predict(X_test)

results["VotingRegressor"] = {
    "MAE": mean_absolute_error(y_test, voting_preds),
    "RMSE": np.sqrt(mean_squared_error(y_test, voting_preds)),
    "R2": r2_score(y_test, voting_preds)
}
```

```python
# Stacking method (advanced ensembling)

estimators = [
    ('rf', models["RandomForest"]),
    ('gb', models["GradientBoosting"]),
    ('svr', models["SVR"]),
    ('knn', models["KNN"])
]

stacking_reg = StackingRegressor(
    estimators=estimators,
    final_estimator=Ridge(),   # meta-model
    passthrough=True
)

stacking_reg.fit(X_train, y_train)
stacking_preds = stacking_reg.predict(X_test)

results["StackingRegressor"] = {
    "MAE": mean_absolute_error(y_test, stacking_preds),
    "RMSE": np.sqrt(mean_squared_error(y_test, stacking_preds)),
    "R2": r2_score(y_test, stacking_preds)
}
```

```python
# Final results
# Mean Absolute error: absolute diff betn actual and predicted
# Root Mean Square Error: square root of the average of squared errors
# R2: explains the varaince in our prediction of weight
results_df = pd.DataFrame(results).T.sort_values(by="R2", ascending=False)
print(results_df)
```

Voting Regressor seems to be better than Stacking regressor. ExtraTrees and RandomForest both performs very well.

```python
results_df = pd.DataFrame(results).T.sort_values(by="R2", ascending=False).reset_index()
results_df = results_df.rename(columns={"index": "Model"})

sns.set(style="whitegrid", font_scale=1.1)
fig, axes = plt.subplots(1, 3, figsize=(18, 8))

# R plot
sns.barplot(x="R2", y="Model", data=results_df, ax=axes[0], color="steelblue")
axes[0].set_title("Model Performance (R)")
axes[0].set_xlim(0, 1)

# MAE plot
sns.barplot(x="MAE", y="Model", data=results_df, ax=axes[1], color="salmon")
axes[1].set_title("Model Performance (MAE)")

# RMSE plot
sns.barplot(x="RMSE", y="Model", data=results_df, ax=axes[2], color="mediumseagreen")
axes[2].set_title("Model Performance (RMSE)")

plt.tight_layout()
plt.show()
```
